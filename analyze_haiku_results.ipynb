```python
output_df = pd.read_csv('output_haiku.csv')
```

```python
def refusal_lambda(x):
    if x.split('=')[1][1] == 'I':
        return True
    else:
        return False
```

```python
output_df['refusal'] = output_df['outputs'].apply(lambda x: refusal_lambda(x))
```

```python
output_df['outputs_beg'] = output_df['outputs'].apply(lambda x: x.split('=')[1][:10])
```
